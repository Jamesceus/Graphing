```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data='/content/drive/MyDrive/PTDataScience/Coding Dojo/Week 4: Visualization for Reporting/Data.xlsx'
sports_df= pd.read_excel(data)
sports_df
```

```python
sports_df.head(10)
```

```python
sports_df.info()
```

```python
plt.plot(sports_df['Sport'], sports_df[2004], color = 'b', marker = '.', label = '2004')
plt.plot(sports_df['Sport'], sports_df[2005], color = 'g', marker = '.', label = '2005')
plt.xticks(rotation = 90)
plt.ylim(bottom=0,top=40)
plt.legend(loc="upper right")
plt.xlabel('Sports', fontsize = 16)
plt.ylabel('TV Ratings', fontsize = 16)
plt.title('Most Watched Sports in America', fontsize = 16)
plt.show()
```

From 2004-2005 the most watched sport in America was Football which dominated TV ratings by twice as much as any other sport.
